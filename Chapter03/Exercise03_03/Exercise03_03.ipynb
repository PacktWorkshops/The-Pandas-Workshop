```python
#
# libraries
#
import pandas as pd
import sqlite3
```

```python
#
# read the current customer table
#
customers = pd.read_sql("select * from Customers", 
                        sqlite3.connect("../datasets/company_database.db"))
customers
```

```python
#
# read the new customer data
#
new_customers = pd.read_csv('../datasets/new_customers.csv')
new_customers
```

```python
#
# add to the Customers table
#
new_customers.to_sql("Customers", 
                     sqlite3.connect('../datasets/company_database.db'), 
                     if_exists = 'append',
                     index = False)
```

```python
#
# now read the entire Customers list
#
customers = pd.read_sql("select * from Customers", 
                        sqlite3.connect("../datasets/company_database.db"))
customers
```
