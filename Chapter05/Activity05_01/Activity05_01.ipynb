```python
#
# required libraries
#
import pandas as pd
#
```

```python
#
# read in the file mushroom.csv
#
mushroom = pd.read_csv('../Datasets/mushroom.csv')
mushroom.head()
#
```

```python
#
# list the column names
#
mushroom.columns
```

```python
#
# create a 3-level index using the class, population, and habitat columns
#
my_index = pd.MultiIndex.from_frame(mushroom[['class', 
                                              'population', 
                                              'habitat']])
```

```python
#
# remove the class, population, and habitat columns
#
mushroom.drop(columns = ['class', 'population', 'habitat'], inplace = True)
#
# now set the index to the MultiIndex
#
mushroom.set_index(my_index, inplace = True)
mushroom.head(10)
#
```
