```python
#
# read the echocardiogram data
#
ECHG_data = pd.read_csv('Datasets\\echocardiogram.csv')
ECHG_data.fillna(9999, inplace = True)
ECHG_data
```

```python
#
# extract survival (convert months to years) to survival
# extract still_alive to alive
# extract age_at_heart_attach to age
#
deaths = ECHG_data.loc[ECHG_data['still_alive'] == 0, :]
survival = deaths['survival'] / 12
age = deaths['age_at_heart_attack'] + survival
#
```

```python
age_data = pd.DataFrame({'survival' : survival,
                         'age' : age})
age_data
```

```python
deaths.reset_index(drop = True, inplace = True)
```

```python
lvdd_hypothesis = pd.DataFrame({'lvdd' : deaths['lvdd'],
                                'survival' : age_data.survival})
lvdd_hypothesis
```

```python
age_data.set_index(deaths.index, drop = True, inplace = True)
lvdd_hypothesis = \
pd.DataFrame({'lvdd' : deaths['lvdd'],
              'survival' : age_data['survival']})
lvdd_hypothesis
```

```python
age_data = pd.DataFrame({'survival' : survival,
                         'age' : age})
lvdd = deaths['lvdd']
lvdd.index = age_data.index
lvdd_hypothesis = pd.DataFrame({'lvdd' : lvdd,
                               'survival' : age_data['survival']})
lvdd_hypothesis
```
