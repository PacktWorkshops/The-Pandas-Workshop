```python
#
# libraries
#
import pandas as pd
```

```python
#
# create the DataFrame with the sale data
#
sales_data = \
pd.DataFrame({'date' : ['2017-03-31', '2017-06-30', '2017-09-30',
                        '2017-12-31', '2018-03-30', '2018-06-30',
                        '2018-09-30', '2019-12-31', '2019-03-31',
                        '2019-06-30', '2019-09-30', '2019-12-31'],
              'sales' : [199190.4, 194356.6, 191611.7, 
                         198918.9, 200163.2, 201510.2, 
                         209749.8, 201897.8, 200098.8, 
                         219340.3, 211542.5, 211729.1]})
sales_data
```

```python
#
# create the month column by slicing from the date strings
#
sales_data['month'] = \
[sales_data.loc[i, 'date'][5:7] for i in sales_data.index]
sales_data.set_index('month', drop = True, inplace = True)
sales_data
```

```python
print('using .iloc with index 3:\n ', sales_data.iloc[3, :])
print('\nusing .loc with index 03:\n', sales_data.loc['03', :])
```

```python
#
# summarize sales by quarter
#
sales_data.groupby('month').mean()
```
