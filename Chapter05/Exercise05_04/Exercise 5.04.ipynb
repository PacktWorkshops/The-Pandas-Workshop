```python
#
# required libraries
#
import pandas as pd
#
```

```python
#
# read in the file mushroom.csv
#
mushroom = pd.read_csv('..\\Datasets\\mushroom.csv')
mushroom
#
```

```python
#
# select the class column, and display how many unique values there are in class
# print the result out in a human readable format
#
print('there are', len(mushroom['class'].unique()), 
      'unique mushroom classes', 
      [mushroom['class'].unique()[i]
       for i in range(len(mushroom['class'].unique()))])
#
```

```python
#
# find all the edible mushrooms and report how many are in the data
#
edibles = mushroom.loc[mushroom['class'] == 'e', :]
print('there are', edibles.shape[0], 'edible mushroom examples')
#
```

```python
#
# assume you only look at the top of the cap
# how many different cases of edibles would you be able to identify?
#
edibles_simple = edibles[['cap-shape', 
                          'cap-surface', 
                          'cap-color']].drop_duplicates()
print('there are', edibles_simple.shape[0], 
      'unique cap combinations in the edibles')
#
```

```python
#
# remove the poisonous mushrooms from the data, storing the result
# in a new data frame, edible_mushrooms
#
print(mushroom.shape)
edible_mushrooms = mushroom.copy()
edible_mushrooms = edible_mushrooms.loc[mushroom['class'] == 'e', :]
print(edible_mushrooms.shape)
#
```

```python
#
# loop over edible_mushrooms, and make a new column 'gill_combos'
# containing the combination of the 'gill' columns separated by '_'
#
gill_cols = ['gill-attachment', 'gill-spacing', 'gill-size', 'gill-color']
edible_mushrooms.gill_combos = ''
for i in range(edible_mushrooms.shape[0]):
    edible_mushrooms.loc[i, 'gill_combos'] = \
    '_'.join([edible_mushrooms.loc[i, gill_cols[j]] for j in range(len(gill_cols))])
#
```

```python
#
# inspect the index
#
edible_mushrooms.index
```

```python
#
# fix the problem by iterating over the index instead of integers
#
edible_mushrooms.gill_combos = ''
for i in edible_mushrooms.index:
    edible_mushrooms.loc[i, 'gill_combos'] = \
    '_'.join([edible_mushrooms.loc[i, gill_cols[j]] 
              for j in range(len(gill_cols))])
#
edible_mushrooms.head()
```

```python
#
# display the first 4 values of the index of edible_mushrooms
#
row_range = 4
list(edible_mushrooms.index)[:row_range]
#
```

```python
#
# print the first 'row_range' odor values
#
# method 1
#
print('method 1--use the index directly\n')
for i in list(edible_mushrooms.index)[:row_range]:
    print(edible_mushrooms.odor[i])
#
# method 2
#
print('method 2--reset the index\n')
edible_mushrooms.reset_index(drop = True, inplace = True)
for i in range(row_range):
    print(edible_mushrooms['odor'][i])
#
```
